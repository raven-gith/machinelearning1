## Apa itu Machine Learning?

> _"Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed."_  
— Arthur Samuel (1959)

> _"A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E."_  
— Tom Mitchell (1997)

Contoh konkret: filter spam email yang belajar dari data email sebelumnya.




## Jenis-jenis Machine Learning

1. **Supervised Learning**  
   - Input data + label (contoh: klasifikasi email spam / bukan spam)

2. **Unsupervised Learning**  
   - Data tanpa label (contoh: segmentasi pelanggan)

3. **Semi-supervised Learning**  
   - Gabungan data berlabel dan tidak berlabel

4. **Reinforcement Learning**  
   - Agen belajar dari lingkungan melalui reward dan punishment


```python

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_blobs

# Supervised dataset
X_sup, y_sup = make_classification(n_features=2, n_redundant=0, n_informative=2,
                                   n_clusters_per_class=1, n_samples=100, random_state=42)

# Unsupervised dataset
X_unsup, _ = make_blobs(n_features=2, centers=3, n_samples=100, random_state=42)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].scatter(X_sup[:, 0], X_sup[:, 1], c=y_sup, cmap='coolwarm')
axs[0].set_title('Supervised Learning (with labels)')

axs[1].scatter(X_unsup[:, 0], X_unsup[:, 1])
axs[1].set_title('Unsupervised Learning (no labels)')

plt.show()
```
